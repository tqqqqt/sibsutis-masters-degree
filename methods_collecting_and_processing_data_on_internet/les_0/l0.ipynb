```python
from google.colab import files

symbols = [ '.', ',', '!', '?', '(', ')', '_', '-', '*', '[', ']', '/', ';', ':', '“', '”', '\n']

# only for google colab
uploaded = files.upload()
for filename in uploaded.keys():
# -------
  res_dict = {}
  with open(filename,'r') as file:
    for line in file:
      clear_line = ''
      for symb in line:
        if symb not in symbols: clear_line=clear_line+symb
      temp_list = clear_line.split()
      for word in temp_list:
        if word.lower() in res_dict: res_dict[word.lower()]+=1
        else: res_dict[word.lower()]=1
  res_dict=sorted(res_dict.items(),key=lambda item : item[1],reverse=True)
  for x in res_dict: print(f'{x[0]} = {x[1]}')
```
